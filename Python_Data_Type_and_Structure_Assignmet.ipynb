Q2. Explain the difference between mutable and immutable data types with examples?

-> In Python, data types are classified as mutable and immutable based on whether their values can be changed after creation.

1. Mutable Data Types

---



  * Mutable data types are those whose values can be modified without changing their memory location.

  * Examples: list, dict, set.

```python
#Example

lst = [1, 2, 3]
lst[0] = 10
print(lst)   # Output: [10, 2, 3]
```

2. Immutable Data Types


---


  * Immutable data types are those whose values cannot be changed once created. Any modification creates a new object in memory.

  * Examples: int, float, string, tuple.

```python
#Example

x = 10
x = 20
print(x)   # Output: 20
```

  | Mutable                       | Immutable                        |
  | ----------------------------- | -------------------------------- |
  | Can be changed after creation | Cannot be changed after creation |
  | Same memory location is used  | New memory location is created   |
  | Examples: list, dict, set     | Examples: int, string, tuple     |

Q10. Explain why strings are immutable in Python?

-> Reasons why strings are immutable are:

---



1. Memory efficiency

* Python can reuse string objects safely.

*  If two variables have the same string value, Python can point both to the same memory location.

2. Performance (fast hashing)

* Strings are often used as dictionary keys.

* Immutability ensures the hash value never changes, which keeps lookups fast and reliable.

3. Thread safety

* Immutable objects cannot be modified by another thread.

* This avoids unexpected changes in multi-threaded programs.

4. Security and reliability

* Prevents accidental modification of data.

* Makes program behavior predictable.

Q14. How does the “in” keyword work differently for lists and dictionaries?

-> The in keyword checks membership, but it works differently for lists and dictionaries in Python.

in with a list

* It checks whether a value (element) exists in the list.

```python
#Example


nums = [10, 20, 30]

print(20 in nums)   # True
print(40 in nums)   # False
```

in with a dictionary

* It checks whether a key exists in the dictionary (not the value).

```python
#Example

data = {"a": 1, "b": 2}

print("a" in data)   # True
print(1 in data)     # False
```

Conclusion:

* The in keyword behaves differently because lists store values, while dictionaries store key–value pairs—so in searches elements in lists and keys in dictionaries for efficient lookup.

Q16.  What is a nested dictionary, and give an example of its use case?

-> Definition

* A nested dictionary is a dictionary where one or more values are themselves dictionaries. It is used to store structured or hierarchical data.

#example

students = {
    "101": {"name": "Arbab", "age": 20, "course": "Python"},
    "102": {"name": "Ali", "age": 21, "course": "Data Science"}
}

Here, each student ID maps to another dictionary containing student details.

```python
#Example

students = {
    "101": {"name": "Arbab", "age": 20, "course": "Python"},
    "102": {"name": "Ali", "age": 21, "course": "Data Science"}
}

print(students["101"])
print(students["102"])
```

Q17.  Describe the time complexity of accessing elements in a dictionary?

-> Time Complexity of Accessing Elements in a Dictionary:

* In Python, dictionary element access using a key (e.g., dict[key]) is very fast because dictionaries are implemented using a hash table.

1. Average Case

* Time Complexity: O(1)

* Accessing a value by its key usually takes constant time, regardless of the dictionary size.

2. Worst Case

* Time Complexity: O(n)

* This can happen rarely due to hash collisions, where many keys end up in the same bucket.

Q18.  In what situations are lists preferred over dictionaries?

-> Lists are preferred over dictionaries in the following situations:

1. When order matters

* Use lists when the sequence of elements is important, such as steps in a process or ordered data.

2. When data is indexed numerically

* Lists are ideal when you want to access elements using position (index) like list[0], list[1].
3. When storing simple collections

* If you only need to store values (not key–value pairs), lists are simpler and more memory-efficient.

4. When duplicate values are needed

* Lists allow duplicate elements, while dictionary keys must be unique.
5. When iterating over all elements

* Lists are better when you mainly loop through data rather than search by key.

Q19. Why are dictionaries considered unordered, and how does that affect data retrieval?

-> Dictionaries are considered unordered because they store data using a **hash table**, where elements are placed based on hash values rather than position; as a result, data retrieval is **very fast (O(1) on average)** using keys, but elements cannot be accessed by index or relied on for order.


Q20.  Explain the difference between a list and a dictionary in terms of data retrieval?

-> **Difference between a List and a Dictionary in terms of data retrieval:**

* **List:** Data is retrieved using a **numeric index** (position). To find a specific value, Python may need to **search through elements**, which can be slower for large lists.

* **Dictionary:** Data is retrieved using a **key**. Python uses **hashing**, so values are accessed **directly and very fast** (average O(1) time), even for large dictionaries.
